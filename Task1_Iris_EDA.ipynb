# **Task 1: Exploring and Visualizing a Simple Dataset **

# Goal: Load, inspect, and visualize the dataset to
#       understand data trends, distributions.
# Dataset: Iris


```python
# Loading the Iris dataset using pandas and convert to csv fromat
import pandas as pd
df = pd.read_csv('iris.csv')
print(df.head())
```

```python
# Summary Statistics
df.info()
print(df.describe())
```

```python
# Visualizing the datset using seaborn and matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

# Scatter plot
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=df)
plt.title('Sepal Length vs Sepal Width')
plt.show()
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.title('Petal Length vs Petal Width')
plt.show()

# Histogram
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

fig, axes = plt.subplots(2, 2, figsize=(10, 7))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.histplot(data=df, x=feature, hue='species', kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {feature}')

plt.suptitle('Feature Distributions by Species')
plt.tight_layout()
plt.show()

# Box plot
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']

fig, axes = plt.subplots(2, 2, figsize=(10, 7))
axes = axes.flatten()

for i, feature in enumerate(features):
    sns.boxplot(data=df, x='species', y=feature, palette='Set2', ax=axes[i])
    axes[i].set_title(feature)

plt.suptitle('Box Plots - Outlier Detection')
plt.tight_layout()
plt.show()
```
